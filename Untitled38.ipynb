```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
from google.colab import files
uploaded = files.upload()
```

```python


file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

print("✅ Data Loaded")
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df = df.drop_duplicates()
df = df.dropna()
print("✅ Data Cleaned")
df.info()
```

```python

print("Total Sales:", df['Sales'].sum())
print("Average Sales:", df['Sales'].mean())
print("Max Sales:", df['Sales'].max())
print("Min Sales:", df['Sales'].min())
```

```python
df_sorted = df.sort_values('Order Date')

plt.figure()
plt.plot(df_sorted['Order Date'], df_sorted['Sales'])
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(10, 6))
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df = df.sort_values('Order Date')
df['Days'] = (df['Order Date'] - df['Order Date'].min()).dt.days
X = df[['Days']]
y = df['Sales']
model = LinearRegression()
model.fit(X, y)
future_days = np.array(range(df['Days'].max()+1, df['Days'].max()+31)).reshape(-1,1)
predictions = model.predict(future_days)
print("📈 Next 30 Days Prediction:")
print(predictions)
```

```python
plt.figure()
plt.plot(df['Days'], df['Sales'], label="Actual Sales")

plt.plot(future_days, predictions, linestyle='dashed', label="Forecast")

plt.legend()
plt.title("Sales Forecast")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```
